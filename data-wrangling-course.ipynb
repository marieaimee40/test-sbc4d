```python
df = pd.read_excel("HDR 2007-2008 Table 27.xlsx", header=6)
df = df.dropna(how="all", axis=1)
df = df.dropna(how="all", axis=0).reset_index(drop=True)
df.head()
```

```python
df.columns #nous permet de connaitre tous les éléments qui sont dans les colonnes
```

```python
#Déposer les lignes du bas des données que nous voulons conserver, jusqu'à la fin (défini comme len(dt))
df = df.drop(np.arange(182,len(df)), axis=0).reset_index(drop=True)
```

```python
MA = ["HDI RANK",
    "COUNTRY",
    "Intentional homicides                 (per 100,000 people)                         2000-04",
    "Reference 1",
    "Prison population     Total                   2007",
    "Reference 2",
    "Prison population     (per 100,000 people)                        2007",
    "Reference 3",
    "Prison population     Female                         (% of total)             2007",
    "Reference 4",
    "Year in which countries have partially or completely abolished the death penalty",
    "Reference 5"
]
```

```python
len(MA)
```

```python
len(df.columns)
```

```python
df.columns=MA
df.columns
```

```python
df.head()
```

```python
#Supprimer toutes les references qui se trouvent dans notre dataframe
df.drop(['Reference 1','Reference 2','Reference 3','Reference 4','Reference 5'],axis='columns',inplace=True)
df.head()
```

```python
# Supprimer les lignes qui contenaient les informations dont nous n'avons plus besoin
# et réinitialiser l'index
# np.arange(2) crée un tableau = [0, 1, 2, 3, 4, 5]
df = df.drop(np.arange(0,2), axis=0).reset_index(drop=True)
df.head(20)
```

```python
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
df["HDI RANK"].index
df[df["HDI RANK"].isin(hdi_categories)].index
```

```python
df = df.drop(df[df["HDI RANK"].isin(hdi_categories)].index, axis=0).reset_index(drop=True)
```

```python
df.loc[:, "HDI RANK"]
```

```python
df.loc[:, "HDI NIVEAU_DEV"] = np.where(df["HDI RANK"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI RANK"] < 71,
                                              "HIGH HUMAN DEVELOPMENT",
                                              "MEDIUM HUMAN DEVELOPMENT"
                                              
                                             )
                                    )
df.sort_values(by="HDI RANK", inplace=True)
df.head()
```

```python
df.tail(30)
```

```python
#Exportons notre fichier sur le format CSV
df.to_csv('HDR 2007-2008 Table 27.csv',index=False) 

#Enregistrons notre fichier en excel avec plusieurs feuilles Données finales, Notes et Sources
with pd.ExcelWriter('HDR 2007-2008 Table 27_KONENATH.xlsx') as writer:
    df.to_excel(writer, sheet_name = 'Données finales', index = False)
    df_NOTES.to_excel(writer, sheet_name = 'NOTES', index = False)
    df_SOURCES.to_excel(writer, sheet_name = 'SOURCES', index = False)
```
